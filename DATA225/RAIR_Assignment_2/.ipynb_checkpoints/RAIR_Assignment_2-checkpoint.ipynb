### Transform the data

````python
def getPositionName(abbr):
    """
        Converts position acronyms to their full names
    """
    positions = {
        'GK':  'Goalkeeper',
        'DR':  'Right Defender',
        'DC':  'Central Defender',
        'DL':  'Left Defender',
        'DMC': 'Defensive Midfielder, Center',
        'AMR': 'Attacking Midfielder, Right',
        'AMC': 'Attacking Midfielder, Center',
        'AML': 'Attacking Midfielder, Left',
        'FW':  'Forward',
        'Sub': 'Substitute',
        'MR':  'Midfielder, Right',
        'MC':  'Midfielder, Center',
        'ML':  'Midfielder, Left',
        'FWR': 'Forward, Right',
        'FWL': 'Forward, Left',
        'DMR': 'Defensive Midfielder, Right',
        'DML': 'Defensive Midfielder, Left',
    }

    return positions.get(abbr, 'Unknown')
````

````python
# Transforming the position columns to their full names
df_appearances['position'] = df_appearances['position'].apply(getPositionName)

# reducing the size of the data set to fit 100 row limit
df_appearances = df_appearances[df_appearances["gameID"].isin(selected_games_list)]

# creating an ID column for the appearances table
df_appearances['ID'] = range(1, len(df_appearances) + 1)
df_appearances.describe()
````

### Create Database tables

````python

cursor.execute('DROP TABLE IF EXISTS teamstats')
cursor.execute('DROP TABLE IF EXISTS teams')

# teams table
sql = ( """
        CREATE TABLE teams
        (
            teamID       int,
            name         varchar(32),
            PRIMARY KEY(teamID)
        )
        """
      )

cursor.execute(sql);


# teamstats table
sql = ("""
        CREATE TABLE teamstats
        (
            ID                 int PRIMARY KEY,
            teamID             int UNIQUE,
            goals              int,
            shots              int,
            shotsOnTarget      int,
            deep               int,
            fouls              int,
            corners            int,
            yellowCards        int,
            redCards           int,
            FOREIGN KEY (teamID) REFERENCES teams(teamID)
        )
        """
      )
cursor.execute(sql);
````

**`teams`** table has `teamID` as a `primary key` and **`teamstats`** table has `teamID` column as a `foreign key` that refers to teams table `teamID` Primary key.

A function to clean (transform) each row:¶
* yellowCards 'NA' ==> 0
* result 'W' ==> 2
* result 'D' ==> 1
* result 'L' ==> 0

Change `result` column to integer values.

````python
def transform(row):
    if row[13] == 'NA':
        row[13] = 0

    _result = row[-1]

    if _result == 'W':
        row[-1] = 2
    elif _result == 'D':
        row[-1] = 1
    elif _result == 'L':
        row[-1] = 0
    else:
        try:
            row[13] = int(yellowCards)  # Convert other numeric values to int
        except ValueError:
            row[13] = 0
````

````python
sql_teams = (   """
              INSERT INTO teams
              VALUES (%s, %s)
              """
          )

sql_teamstats = (   """
               INSERT INTO teamstats
               VALUES (%(ID)s, %(teamID)s, %(goals)s, %(shots)s, %(shotsOnTarget)s, %(deep)s, %(fouls)s, %(corners)s, %(yellowCards)s, %(redCards)s)
               """
           )
````

## League and shots table

````python
cursor.execute('DROP TABLE IF EXISTS league')
cursor.execute('DROP TABLE IF EXISTS shots')
````

````python
sql_league = ( """
        CREATE TABLE league
        (
            leagueID     int,
            name   varchar(32),
            understatNotation  varchar(32),
            PRIMARY KEY(leagueID)
        )
        """
      )

sql_shots = ( """
        CREATE TABLE shots
        (
            ID int PRIMARY KEY,
            gameID     int,
            shooterID   int,
            assisterID   int,
            minute   int,
            situation  varchar(32),
            lastAction  varchar(32),
            shotType  varchar(32),
            shotResult  varchar(32),
            xGoal double,
            positionX    double,
            positionY    double
        )
        """
      )


cursor.execute(sql_league)
cursor.execute(sql_shots)
````

````python
def transform(row):
    if row[2] == 'NA':
        row[2] = 0
````

````python
sql_all = (   """
              INSERT INTO league
              VALUES (%s, %s, %s)
              """
          )

first = True
with open('./data/leagues.csv', newline='') as csv_file:
    data = csv.reader(csv_file, delimiter=',', quotechar='"')
    for row in data:
        if first == False:
            transform(row)
            cursor.execute(sql_all, row)
        first = False
conn.commit()
````

````python
sql_all = (   """
              INSERT INTO shots
              VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
              """
          )

first = True
with open('./data/shots.csv', newline='') as csv_file:
    data = csv.reader(csv_file, delimiter=',', quotechar='"')
    idx = 0
    for row in data:
        if first == False and int(row[0]) in selected_games_list:
            idx += 1
            transform(row)
            row.insert(0,idx)
            cursor.execute(sql_all, row)
        first = False
conn.commit()
````

### 5 queries with joins

````python
# example of one to one relationship
# each team has its own stats
_, df = dataframe_query(conn, """
SELECT *
FROM rair_football.teams t
join rair_football.teamstats ts on t.teamID = ts.teamID
""")
df
````

````python
# example of many to many relationship
# many players can make shots in many games
_, df = dataframe_query(conn, """
SELECT playerID, games.gameID FROM rair_football.players, rair_football.shots, rair_football.games
Where games.gameID = shots.gameID
AND shooterID = playerID
order by playerID
limit 10
""")
df
````

````python
# example of one to many relationship
# one player can make many shots
_, df = dataframe_query(conn, """
SELECT name, shotResult, gameID
FROM rair_football.players, rair_football.shots
WHERE name = 'Marc Pugh'
AND playerID = shooterID
""")
df
````

````python
# example of one to many relationship
# one league can have many games
_, df = dataframe_query(conn, """
select * from league l
join games g on g.leagueID = l.leagueID
""")
df
````

````python
# example of many to many relationship
# many player can appear in many games
_, df = dataframe_query(conn, """
select g.gameID, g.season, g.date, p.playerID, p.name from games g
join appearances a on a.gameID = g.gameID
join players p on p.playerID = a.playerID
""")
df
````
