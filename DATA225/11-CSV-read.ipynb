# Read a CSV File

```python
import csv
```

#### We will get a **list of values** for each row of the CSV file.

```python
with open('TitanicSurvival.csv', newline='') as csv_file:
    data = csv.reader(csv_file, delimiter=',', quotechar='"')
    
    
     
    for row in data:
        print(row)
```
